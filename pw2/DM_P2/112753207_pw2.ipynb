```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import cluster
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# 讀取資料
marketData = pd.read_csv("./data/marketData.csv")
```

```python
marketData
```

```python
###第一題

# 挑出女性客戶的資料
female_customers = marketData[marketData['Gender'] == 'Female'][['Age', 'Spending Score (1-100)']]

# 使用K-means分群方法，設定K=2，random_state為隨機初始值(通常設定為42)
kmeans = KMeans(n_clusters=2, random_state=42).fit(female_customers)

# 取得分群的中心點
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=['Age', 'Spending Score (1-100)'])

# 根據Age數值升冪排序
cluster_centers_sorted = cluster_centers.sort_values(by='Age')

# 列印出兩個群組的中心點
print("兩個群組的中心點：")
print(cluster_centers_sorted)
```

```python
###第二題

# 嘗試不同的K值，並計算對應的群內平方和
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(female_customers)
    inertia.append(kmeans.inertia_)

# 繪製Elbow 方法的圖表，群內樣本與群心之間的距離平方和（Inertia）
# 當K值達到一定程度時，Inertia的下降速度會急劇減緩，這個轉折點通常被稱為"肘部"（Elbow）
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='-', color='b')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.xticks(range(1, 11))
plt.grid(True)
plt.savefig('第二題.jpg')
plt.show()
```

```python
###第三題

# 挑出女性客戶的資料
female_customers = marketData[marketData['Gender'] == 'Female'][['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# 使用K-means分群方法，設定K=3，random_state為隨機初始值(通常設定為42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(female_customers)

# 取得分群的中心點
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])

# 根據Age數值升冪排序
cluster_centers_sorted = cluster_centers.sort_values(by='Age')

# 列印出三個群組的中心點
print("三個群組的中心點：")
print(cluster_centers_sorted)
```

```python
###第四題

# 嘗試不同的K值，並計算對應的群內平方和
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(female_customers)
    inertia.append(kmeans.inertia_)

# 繪製Elbow 方法的圖表，群內樣本與群心之間的距離平方和（Inertia）
# 當K值達到一定程度時，Inertia的下降速度會急劇減緩，這個轉折點通常被稱為"肘部"（Elbow）
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='-', color='b')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.xticks(range(1, 11))
plt.grid(True)
plt.savefig('第四題.jpg')
plt.show()
```

```python
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(female_customers)
    score = silhouette_score(female_customers, kmeans.labels_)
    silhouette_scores.append(score)

# 繪製Silhouette分數的圖表
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='b')
plt.xlabel('K')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.xticks(range(2, 11))
plt.grid(True)
plt.savefig('第四題方法二Silhouette_analysis.jpg')
plt.show()
```
